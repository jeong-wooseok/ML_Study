# Contents Based Filtering_tmdb5000

## 영화정보 전처리

```python
import numpy as np
import pandas as pd
import warnings; warnings.filterwarnings('ignore')

movies = pd.read_csv('tmdb_5000_movies.csv')
movies
```

```python
movies_df = movies[['id','title','genres','vote_average','vote_count','popularity',
'keywords','overview']]
movies_df
```

```python
pd.set_option('max_colwidth',100) # 한열당 100자까지 보도록 설정
movies_df[['keywords','genres']] 
```

```python
## 매우 중요 :  ast 모듈의 liter_eval()을 사용하면 문자열을 텍스트가 아닌 list[dict1,dict2]객체를 부를 수 있음
from ast import literal_eval
movies_df['genres'] = movies_df['genres'].apply(literal_eval)
movies_df['keywords'] = movies_df['keywords'].apply(literal_eval)

# 단순한 리스트 형태로 전처리
movies_df['genres'] = movies_df['genres'].apply(lambda x : [y['name'] for y in x])
movies_df['keywords'] = movies_df['keywords'].apply(lambda x : [y['name'] for y in x])

movies_df[['keywords','genres']]
```

## 장르콘텐츠 유사도 측정

```python
from sklearn.feature_extraction.text import CountVectorizer

# CountVectorizer 적용하기 위해 공백문자로 word단위가 구분되는 문자열 변환

movies_df['genres_literal'] = movies_df['genres'].apply(lambda x : (' ').join(x))
count_vect = CountVectorizer(min_df=0, ngram_range=(1,2))
genre_mat = count_vect.fit_transform(movies_df['genres_literal'])
genre_mat
# dir(genre_mat)
```

276개의 장르 피처를 지닌 4803행의 매트릭스 생성 <br>
이제 cosine유사도로 거리를 측정한다

```python
from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(genre_mat, genre_mat)
genre_sim
```

```python
# 점수는 계산 되었으니, 각 레이블 별로 높은 순위부터 장르 정렬할 필요 있음 : 넘파이의 argsort() 사용
genre_sim_sorted_idx = genre_sim.argsort()[:,::-1] # 내림차순으로 우선정렬
genre_sim_sorted_idx
```

## 순위 추천함수 불러오기

```python

def find_sim_movies(df, sorted_idx, title_name, top_n = 10):
    # 인자로 입력된 movie_df에서 'title' 컬럼이 입력된 title_name 값인 DataFrame추출
    title_movie = df[df['title'] == title_name]

    # title_name을 가진 df의 index객체를 ndarray로 반환하고
    # sorted_idx 입력된 genre_sim_sorted_idx에서 유사도 순으로 top_n의 idx추출
    title_idx = title_movie.index.values
    similar_idx = sorted_idx[title_idx, : (top_n)]

    print(similar_idx)
    similar_idx = similar_idx.reshape(-1)
    return df.iloc[similar_idx] 
```

```python
similar_movies = find_sim_movies(movies_df, genre_sim_sorted_idx, 'The Godfather')
```

듣보잡 영화도 장르만으로 가져오다보니, 많이 끌고 온 것 확인. 어쩌지? <br>
가중치가 부여된 평점을 활용해서 가중치를 주고자 함 <br>

- 개별영화 평점투표 횟수 : v
- 평점부여 최소투표 횟수 : m (투표횟수에 직접 가중치를 조정하는 역할을 함, m을 높이면 평점투표횟수가 많은 영화에 더 많은 가중평점 부여)
- 개별영화 평균평점 : R
- 전체영화 평균평점 : C
<br><br>
- 가중평점 = (v/(v+m)) * R + (m/(v+m)) * C



```python
C = movies_df['vote_average'].mean()
m = movies_df['vote_count'].quantile(0.6) # 전체투표중에서 상위 60%에 해당하는 횟수를 기준으로 정함

def weighted_vote_average(record):
    v = record['vote_count']
    R = record['vote_average']

    return ((v/(v+m)) * R) + ((m/(v+m)) * C)

movies_df['weighted_vote'] = movies_df.apply(weighted_vote_average, axis=1)
```

```python
movies_df.sort_values('weighted_vote', ascending=False)[:10] # weighted_vote 상위 10개 영화 확인
```

```python

def find_sim_movies_weighted(df, sorted_idx, title_name, top_n = 10):
    # 인자로 입력된 movie_df에서 'title' 컬럼이 입력된 title_name 값인 DataFrame추출
    title_movie = df[df['title'] == title_name]

    # title_name을 가진 df의 index객체를 ndarray로 반환하고
    # sorted_idx 입력된 genre_sim_sorted_idx에서 유사도 순으로 top_n의 idx추출
    title_idx = title_movie.index.values
    similar_idx = sorted_idx[title_idx, : (top_n*2)] # 두배 더 뽑고
    similar_idx = similar_idx.reshape(-1)
    similar_idx = similar_idx[similar_idx != title_idx] # 기준영화명은 제외

    return df.iloc[similar_idx].sort_values('weighted_vote', ascending = False)[:top_n] # weighted_vote로 재정렬
```

```python
similar_movies_weighted = find_sim_movies_weighted(movies_df, genre_sim_sorted_idx, 'The Godfather')
similar_movies_weighted
```
